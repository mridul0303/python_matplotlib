```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# line_plot

```python
x=[1,2,3,4,5,6,7]
y=[23,46,34,33,27,43,23]
```

```python
plt.xlabel("day")
plt.ylabel("temperature")
plt.title("Weather")
plt.plot(x,y,color="green",linewidth="5",linestyle="dotted")
```

```python
plt.plot(x,y,"Db ")#D=diamond
#space given for no line
```

```python
plt.plot(x,y,color="#888888",marker="*",markersize=20,alpha=0.5)
#"#888888 represnts hexa value for rbg combination
#alpha handles transparency range 0 to 1
```

```python
day=[1,2,3,4,5,6,7]
max_temp=[34,56,39,44,42,47,38]
min_temp=[23,46,34,33,23,33,23]
avg_temp=[30,50,36,39,31,38,29]
```

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")
plt.plot(day,max_temp,label="Max")
plt.plot(day,avg_temp,label="avg")
plt.plot(day,min_temp,label="Min")
plt.legend(loc="best",shadow=True, fontsize="small")#shadow for label/legend presentation
plt.grid()
```

# Bar Graph

```python
x=["mon","tues","wed","thurs","fri","sat","sun"]
y=[23,46,34,33,23,33,23]
h=[15,23,21,22,13,23,14]
```

```python
plt.xlabel("Days")
plt.ylabel("temp")
plt.bar(x,y)
```

```python
ypos=np.arange(len(x))
ypos
```

```python
plt.bar(ypos,y)
```

```python
plt.xticks(ypos,x)
plt.xlabel("Days")
plt.ylabel("temp")
plt.bar(ypos-0.2,y,label="temp",width=0.4)
plt.bar(ypos+0.2,h,label="humdity",width=0.4)
plt.legend()
```

```python
plt.yticks(ypos,x)
plt.xlabel("temp")
plt.ylabel("days")
plt.barh(ypos-0.2,y,label="temp")
plt.barh(ypos+0.2,h,label="humdity")
plt.legend()
```

# Histograms

```python
y=[13,46,24,3,27,43,35]#temp
#wont need categorical value as it defines occurences
```

```python
plt.hist(y,bins=3,rwidth=.90)
```

```python
plt.hist(y,bins=[10,20,30,40,50])
```

```python
h=[15,23,56,42,37,19,28]
plt.hist([y,h],bins=[10,20,30,40,50],color=['g','r'],label=["temp","humidity"],alpha=0.5)
plt.legend(loc="best",shadow=True)
```

```python
plt.hist([y,h],bins=[10,20,30,40,50],color=['g','r'],label=["temp","humidity"],alpha=0.5,orientation='horizontal')
plt.legend(loc="best",shadow=True)
```

# Pie Chart

```python
x=["mon","tues","wed","thurs","fri","sat","sun"]
y=[23,46,34,65,33,13,53]
plt.pie(y,labels=x)
```

```python
plt.axis="equal" #alignmnet
plt.pie(y,labels=x,radius=2,autopct="%0.2f%%",shadow=True,explode=[0,0,0,0,0.3,0,0.7],startangle=45)
plt.show() #removes output data
```

# Savefig

```python
plt.pie(y,labels=x,radius=2,autopct="%0.2f%%",shadow=True,explode=[0,0,0,0,0.3,0,0.7],startangle=45)
plt.savefig("temp_chart2.png",bbox_inches="tight",pad_inches=3,facecolor="white",transparent=True)
```
